```python
import pandas as pd
from preprocess.library import *
```

```python
en_val_y = pd.read_csv("preprocess/en_val_y.csv")
y_valid = en_val_y.iloc[:,1:]
```

```python
cv_ovr_none_cat = pd.read_csv("cv_ovr_none_cat.csv")
cv_ovr_rs_lgbm = pd.read_csv("cv_ovr_rs_lgbm.csv")
```

```python
ds = np.array([])

for i in range(100):
    SCDC2_predict = balancing_auroc_lift(cv_ovr_none_cat, cv_ovr_rs_lgbm, (i+1)/100, 1-((i+1)/100))
    ds = np.append(ds, target_score(np.array(SCDC2_predict), y_valid, en_val_y, [3,4,7]), axis=0)
    print(target_score(np.array(SCDC2_predict), y_valid, en_val_y, [3,4,7]))
```

```python
print(ds.reshape(100,2))
```

```python
ds = pd.DataFrame(ds.reshape(100,2))
```

```python
ds.iloc[:,0] = ds.iloc[:,0]*10
```

```python
ds.columns = ['AUROCx10', 'LIFT']
```

```python
ds.plot()
```

```python
ds[40:60].plot()
```

```python
ds.to_csv("data.csv", index=False)
```

```python
SCDC2_predict = balancing_auroc_lift(cv_ovr_none_cat, cv_ovr_rs_lgbm, 0.5, 0.5)
print(totality_score(SCDC2_predict, y_valid, en_val_y))
print(target_score(np.array(SCDC2_predict), y_valid, en_val_y, [3,4,7]))
print(target_score(np.array(SCDC2_predict), y_valid, en_val_y, [1,2,5,6,8,9,10]))
```

```python
SCDC2_predict = balancing_auroc_lift(cv_ovr_none_cat, cv_ovr_rs_lgbm, 0.48, 0.52)
print(target_score(np.array(SCDC2_predict), y_valid, en_val_y, [3,4,7]))
```

```python
SCDC2_predict = balancing_auroc_lift(cv_ovr_none_cat, cv_ovr_rs_lgbm, 0.5, 0.5)
print(target_score(np.array(SCDC2_predict), y_valid, en_val_y, [3,4,7]))
```
